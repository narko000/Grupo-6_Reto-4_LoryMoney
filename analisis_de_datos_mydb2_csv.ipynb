```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/mydb2.csv')

# temperatura media por server
temp_media = df.groupby('idServidor')['Temperatura'].mean()

plt.figure(figsize=(8,4))
temp_media.plot(kind='bar', color='tomato')
plt.title('Temperatura media por servidor')
plt.xlabel('Servidor')
plt.ylabel('Temperatura (°C)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Cconsumo energetico por servidor
consumo = df.groupby('idServidor')['ConsumoEnergetico'].sum()

plt.figure(figsize=(8,4))
consumo.plot(kind='bar', color='steelblue')
plt.title('Consumo energético total por servidor')
plt.xlabel('Servidor')
plt.ylabel('Consumo (kW)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# estadisticas generales
print(df[['Temperatura','ConsumoEnergetico','EmisionesCO2']].describe())
```
